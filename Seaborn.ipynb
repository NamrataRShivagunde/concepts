```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.set_style("whitegrid")
```

```python
g = sns.lmplot(x = "tip",
              y = "total_bill",
              data = tips,
              aspect =2)
```

```python
g = (g.set_axis_labels("Tip","Total bill(USD)"). set(xlim=(0,10), ylim =(0,100)))
```
